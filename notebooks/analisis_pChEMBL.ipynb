```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("/content/drive/MyDrive/Bioinformática/TFM/Final/pChEMBL/dataset_with_pchembl.csv")
```

```python
# Recuento de valores NaN
nan_count = data['pChEMBL'].isna().sum()

# Recuento valores pChEMBL
pchembl_count = data['pChEMBL'].notna().sum()

print(f"Cantidad de valores pChEMBL: {pchembl_count}")
print(f"Cantidad de valores NaN en pChEMBL: {nan_count}")
```

```python
# Resumen estadístico sin tener en cuenta valores 'Nan'
summary = data['pChEMBL'].describe()
print(summary)
```

```python
# Histograma de distribución de valores pChEMBL
plt.figure(figsize=(8, 6))
sns.histplot(data['pChEMBL'], kde=True, bins=30, color='blue')
plt.title("Distribución de pChEMBL")
plt.xlabel("pChEMBL")
plt.ylabel("Frecuencia")
plt.show()
```

```python
# Boxplot de valores pChEMBL
plt.figure(figsize=(8, 6))
sns.boxplot(x=data['pChEMBL'], color='green')
plt.title("Boxplot de pChEMBL")
plt.xlabel("pChEMBL")
plt.show()
```

```python
# Heatmap de valores faltantes
plt.figure(figsize=(10, 6))
sns.heatmap(data.isna(), cbar=False, cmap='viridis')
plt.title("Mapa de valores faltantes")
plt.show()
```
